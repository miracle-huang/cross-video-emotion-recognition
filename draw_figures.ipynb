```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import simps

# 模拟一段PSD曲线
freqs = np.linspace(10, 40, 7)
psd = np.array([2, 5, 8, 6, 4, 3, 1])

# 选取频段
band_idx = (freqs >= 14) & (freqs <= 31)
freqs_band = freqs[band_idx]
psd_band = psd[band_idx]

plt.plot(freqs, psd, 'o-', label='PSD curve')
plt.fill_between(freqs_band, psd_band, color='orange', alpha=0.5, label='Band area (Simpson)')
plt.xlabel('Frequency (Hz)')
plt.ylabel('PSD')
plt.title('Simpson Integration of PSD in Selected Band')
plt.legend()
plt.show()
```
